```python
import pandas as pd
df = pd.read_csv("../data/Computer_Science_curriculum.csv", index_col=0)

concepts = []
for idx, row in df.iterrows():
	course_name, course_content = row["course_name"], row["program_content"]
	if pd.isna(course_content): 
		concepts.append('')
		continue
	isOk = False
	while isOk == False:
		try:
			response = client.models.generate_content(
						model=model,
						contents=create_prompt(course_name, course_content),
			)
			concepts.append(response.text)
			isOk = True
			if idx % 10 == 0: print(f"Idx: {idx}")
		except Exception as e:
			print(f"{e}")
			time.sleep(45)
```
