```python
import pandas as pd


def generate_concepts(prompt_template: str, case_name: str, gt_file="gt.csv"):
	df = pd.read_csv(gt_file, index_col=0)
	concepts = []
	cum_time = 0

	for idx, row in df.iterrows():
		course_name, course_content = row["course_name"], row["program_content"]
		if pd.isna(course_content): 
			concepts.append('')
			continue
		isOk = False
		while isOk == False:
			try:
				start = time.time()
				response = client.models.generate_content(
							model=model,
							contents=create_prompt(prompt_template, course_name, course_content),
				)
				cum_time += time.time() - start
				concepts.append(response.text)
				isOk = True

				# print(f"Idx: {idx}")
			except Exception as e:
				# print(f"{e}")
				time.sleep(30)
			
	df[case_name] = concepts
	df.to_csv(case_name+".csv")
	# print(f"Avg time per generation: {cum_time/len(concepts)}")
	return cum_time/len(concepts)
```

```python
from fuzzywuzzy import fuzz
import re
from sentence_transformers import SentenceTransformer, util
import numpy as np

def normalize_concepts(concept_str):
    if isinstance(concept_str, str):
        return [c.strip().lower() for c in concept_str.split(';') if c.strip()]
    else:
        return []

def remove_brackets(text):
    return re.sub(r"\s*\([^)]*\)", "", text).strip()

def fuzzy_match_with_partial(gt_concepts, gen_concepts, threshold=85, partial_threshold=0.6):
    matches = []
    partial_matches = []
    unmatched_gt = []
    used_indexes = set()

    # Precompute normalized concepts (no brackets)
    norm_gen = [remove_brackets(g.lower()) for g in gen_concepts]
    norm_gt = [remove_brackets(g.lower()) for g in gt_concepts]

    # --- First Pass: Fuzzy matching (ignoring brackets)
    for idx_gt, gt in enumerate(gt_concepts):
        norm_gt_concept = norm_gt[idx_gt]
        best_match = None
        best_score = 0
        best_index = -1

        for idx_gen, gen in enumerate(gen_concepts):
            if idx_gen in used_indexes:
                continue
            norm_gen_concept = norm_gen[idx_gen]

            score = fuzz.token_sort_ratio(norm_gt_concept, norm_gen_concept)

            if score > best_score:
                best_score = score
                best_match = gen
                best_index = idx_gen

        if best_score >= threshold:
            matches.append((gt, best_match, best_score))
            used_indexes.add(best_index)
        else:
            unmatched_gt.append((gt, norm_gt_concept))  # keep both versions

    unmatched_gen = [gen_concepts[i] for i in range(len(gen_concepts)) if i not in used_indexes]

    # --- Second Pass: Partial matches (ignoring brackets)
    still_unmatched = []
    norm_unmatched_gen = [remove_brackets(g.lower()) for g in unmatched_gen]
    unmatched_to_remove = []

    for original_gt, norm_gt_concept in unmatched_gt:
        matched = False
        gt_tokens = set(norm_gt_concept.split())

        for i, norm_gen_concept in enumerate(norm_unmatched_gen):
            gen_tokens = set(norm_gen_concept.split())
            token_overlap = len(gt_tokens & gen_tokens) / len(gt_tokens | gen_tokens)

            if norm_gt_concept in norm_gen_concept or norm_gen_concept in norm_gt_concept or token_overlap >= partial_threshold:
                partial_matches.append((original_gt, unmatched_gen[i], f"token overlap: {token_overlap:.2f}"))
                matched = True
                unmatched_to_remove.append(unmatched_gen[i])
                break
										

        if not matched:
            still_unmatched.append(original_gt)

    return {
        'matches': matches,
        'partial_matches': partial_matches,
        'missing_gt': still_unmatched,
        'additional_gen': [x for x in unmatched_gen if x not in unmatched_to_remove]
    }

embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
def semantic_match(gt_concepts, gen_concepts, threshold=0.7):
    # Encode all phrases
    gt_embeddings = embedding_model.encode(gt_concepts, convert_to_tensor=True)
    gen_embeddings = embedding_model.encode(gen_concepts, convert_to_tensor=True)

    matches = []
    unmatched_gt = []
    unmatched_gen = set(range(len(gen_concepts)))  # index tracking

    for i, gt_emb in enumerate(gt_embeddings):
        similarities = util.cos_sim(gt_emb, gen_embeddings)[0]
        best_idx = int(similarities.argmax())
        best_score = float(similarities[best_idx])

        if best_score >= threshold and best_idx in unmatched_gen:
            matches.append((gt_concepts[i], gen_concepts[best_idx], best_score))
            unmatched_gen.remove(best_idx)
        else:
            unmatched_gt.append(gt_concepts[i])

    unmatched_gen_phrases = [gen_concepts[i] for i in unmatched_gen]

    return {
        "matches": matches,
        "missing_gt": unmatched_gt,
        "additional_gen": unmatched_gen_phrases
    }
```

```python
import pandas as pd

def experiment(experiment_name:str, prompt_name, reps=10):
	for rep in range(reps):
		avg_time=generate_concepts(prompt_name, experiment_name)
		results = []

		gt = pd.read_csv(f"{experiment_name}.csv", index_col=0)

		TP = 0
		FP = 0
		FN = 0

		for idx, row in gt.iterrows():
						gt_list = normalize_concepts(row['educational_concepts'])
						gen_list = normalize_concepts(row[experiment_name])
						
						res = semantic_match(gt_list, gen_list, 0.7)
						
						results.append({
										'course_name': row['course_name'],
										'matches': res['matches'],
										'missing_concepts': res['missing_gt'],
										'additional_concepts': res['additional_gen']
						})
						
						TP += len(res['matches'])
						FP += len(res['additional_gen'])
						FN += len(res['missing_gt'])

						# print("\n", row['course_name'])
						# for m in res['matches']:
						# 				print(f"✔ {m[0]} ↔ {m[1]} (score: {m[2]})")

						# print("❌ Missing:", res['missing_gt'])
						# print("➕ Additional:", res['additional_gen'])
			
		precision = TP / (TP + FP) if (TP + FP) > 0 else 0.0
		recall = TP / (TP + FN) if (TP + FN) > 0 else 0.0
		f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0.0

		print(f"{precision:.2f},{recall:.2f},{f1:.2f},{avg_time:.2f}")
		time.sleep(20)
```
